## Ludolphine number project

As a first application we want to determine the Ludolphine number, better known as $\pi$ numerically.

[Ludolph](http://en.wikipedia.org/wiki/Ludolph_van_Ceulen) spent a major part of his life calculating the numerical value of the mathematical constant $\pi$, using essentially the same methods as those employed by Archimedes some seventeen hundred years earlier. Eventually he determined the first 35 decimals of the [_Ludolphine_ number](http://en.wikipedia.org/wiki/Pi):
$$ 3.14159265358979323846264338327950288...$$


### A numerical approach

We want to calculate $\pi$ without using any trigonometric function, just the function 

$$ y = \sqrt{(1 - x^2)} $$ 

In order to find a numerical solution with must follow the following steps:

* Concept: what is the basic strategy of your approach. This would often involve some math, sketches and diagrams representing the ideas.

* The all important question: do I solve equations and a modeling approach that is appropriate for the task. Too realistic and the program will cost too much to build and/or run. Too simplistic and the results will not be very meaningful.
* Algorithm: a detailed and usually language-agnostic list of steps a computer would have to proceed through in order to implement the concept.
* Implementation: a language-specific version of the algorithm that runs on some hardware and produces quantitative output.
* Testing and debugging: this usually the most neglected (at least by beginners) and yet most time-consuming and ultimately important step. Only if you can convince yourself that the program does exactly what you intended it to do is there hope that it may be useful.
