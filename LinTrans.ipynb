## Linear transformations
Imagine we have 2D set of points (or any object). We can use matrices to express linear transformations for such point sets, such as reflection, rotation or scaling, see [Examples of linear transformation matrices](https://en.wikipedia.org/wiki/Linear_map). Let's implement a framework to experiment with such linear maps. 

```python
%pylab nbagg
```

```python
n=100
points=[[cos(2*pi*t/n),sin(2*pi*t/n)] for t in xrange(n)]
```

```python
points
```

```python
array(points).T[0]
```

```python
x=array(points).T[0]; y=array(points).T[1]
```

```python
ifig=1;close(ifig);figure(ifig,figsize=(5,5))
plot(x,y,'rh')
```

```python
x=array(points).T[0]; y=array(points).T[1]
```

```python
x
```

```python
T1 = matrix([[1,0],[0,2]])
```

```python
vecs=array(points)
```

```python
T1.dot(vecs[3]).A1
```

```python
trans_vec=T1.dot(vecs[0]).A1
```

```python
trans_vec
```

```python
len(vecs)
```

```python
# transformation:
trans_vec = []
for i in range(len(vecs)):
    this_tv = T1.dot(vecs[i]).A1
    trans_vec.append(this_tv)
```

```python
xt=array(trans_vec).T[0]; yt=array(trans_vec).T[1]
```

```python
ifig=1;figure(ifig)
plot(xt,yt,'o')
```
